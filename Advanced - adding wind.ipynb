```python
##### Wind parameters #####
vwindfactor = 0.1
windbeta = 4.0
alpha =  1.0
Mdot = 1e-4
#fRmax =10.0
logRmax = 15
Mdot = Mdot*physcons.MSUN/physcons.YR

#--------------------------------------------------------------------------------------------
list_IIL = ['z014-10-0.3-0', 'z014-10-0.3-0.6', 'z014-10-0.6-2.8', 'z014-10-0.8-2.6', 'z014-10-0.9-2.6', 'z014-12-0.1-0', 'z014-12-0.1-0.6', 'z014-12-0.4-2.8', 'z014-12-0.5-2', 'z014-12-0.5-2.8', 'z014-12-0.6-2', 'z014-12-0.6-2.8', 'z014-12-0.9-2.6', 'z014-15-0.9-2.8', 'z014-20-0.1-3.2', 'z014-25-0.8-3.4', 'z014-25-0.9-3.4', 'z014-8-0.1-0', 'z014-8-0.1-0.4', 'z014-9-0.1-0', 'z014-9-0.1-0.4', 'z014-9-0.1-2.8', 'z014-9-0.8-2.6', 'z014-9-0.9-2.6']
for name in list_IIL:#MODEL:
    abn_ = pd.read_csv(rundir+name+'.abn', header = None, sep = '\s+').values
    hyd_ = pd.read_csv(rundir+name+'.hyd',header = None, skiprows = 1, sep = '\s+').values
    Nwind = int(100*(logRmax-np.log10(hyd_[-1][2]))/1.5)
    
##### Header exchange, execute after wind has added #####
    '''FF = open(rundir+name+'/'+name+'.hyd', 'r')
    table = FF.readlines()
    head = [float(i) for i in table[0].split()]
    table[0] = '{0:9.1e}{1:4d}{2:13.5e}{3:13.5e}{4:13.4e}\n'.format(0.,len(abn_), head[2], head[3], head[4])
    FF = open(rundir+name+'/'+name+'.hyd', 'w')
    FF.writelines(table)
    print(table[:2])'''
#########################################################
    struct = hyd_[-1]
    MH_init = np.dot(hyd_[:,1], abn_[:,4])
    #print(name, '\n','Radius={0:.4e}cm,'.format(struct[2]), 'rho={0:.3e},'.format(struct[3]),\
     #     ' Mass={0:.4f}Msun,'.format(struct[-2]) ,'H_mass={0:.4f}Msun'.format(MH_init))
    
    Mstar = hyd_[-1][6]*physcons.MSUN
    Rstar = hyd_[-1][2]
    Rhostar = hyd_[-1][3]
    Tstar = hyd_[-1][4]
    Lstar = 4*physcons.PI*Rstar**2*physcons.SIG*Tstar**4
    vesc = (2*Mstar*physcons.G/Rstar)**0.5
    vinf = 10e5
    v0 = Mdot/(4.*physcons.PI*Rstar**2*Rhostar)
    print(v0, vinf, Rstar)
    
    dlogR = (logRmax-np.log10(hyd_[-1][2]))/Nwind
    nzone_old = len(abn_)
    nzone_new = len(abn_)+Nwind
    
    print(vesc, dlogR, nzone_new)
    
    Rho_,R_ ,T_= [],[],[]
    xm ,logRstar = hyd_[-1][6], np.log10(hyd_[-1][2])
    
    HYD = open(rundir+name+'.hyd', 'a')
    ABN = open(rundir+name+'.abn', 'a')
    
    for i in range(Nwind):
        rr = 10**(logRstar+dlogR)
        vwind = v0 +  (vinf-v0)*(1.0 - Rstar/rr)**windbeta
        Rho_new = Mdot/(4*physcons.PI*rr**2*vwind)
        Rho_.append(Rho_new)
        T_new = (Lstar/(4*physcons.PI*rr**2*physcons.SIG))**0.25
        T_.append(T_new)
        U_new = vwind
        H_new = abn_[-1][4]
        He_new = abn_[-1][5]
        C_new = abn_[-1][6]
        N_new = abn_[-1][7]
        O_new = abn_[-1][8]
        Ne_new = abn_[-1][9]
        Mg_new = abn_[-1][11]
        Si_new = abn_[-1][13]
        S_new = abn_[-1][14]
        Ar_new = abn_[-1][15]
        Ca_new = abn_[-1][16]
        Fe_new = abn_[-1][17]
        Ni56_new = abn_[-1][19]
        xm_new = xm+4*physcons.PI*rr**2*Rho_[i]*(rr - 10.**logRstar)/physcons.MSUN
        dm_new = xm_new - xm
        xm += dm_new
        logRstar+=dlogR
        R_.append(logRstar)
        if xm_new == 0:
            print("Warning")
        if i == Nwind-1:
            print(name,"WM =",xm - hyd_[-1][6])
            print("Total mass",xm)
##### To prevent overlapping of contents #####
        if hyd_[-1][5] == 0: 
            hyd_line = ' {0:4d}{1:13.4e}{2:13.4e}{3:13.4e}{4:13.4e}{5:13.4e}{6:20.12e}{7:13.4e}\n'\
            .format(nzone_old+1+i,dm_new, rr, Rho_new, T_new, U_new, xm_new, 0.0)
            abn_line = '{0:4d}{1:11.3e}{2:11.3e}{3:11.3e}{4:11.3e}{5:11.3e}{6:11.3e}{7:11.3e}{8:11.3e}{9:11.3e}{10:11.3e}\
        {11:11.3e}{12:11.3e}{13:11.3e}{14:11.3e}{15:11.3e}{16:11.3e}{17:11.3e}{18:11.3e}{19:11.3e}\n'\
            .format(nzone_old+1+i, 0.0, 0.0, 0.0, H_new, He_new, C_new, N_new, O_new, Ne_new, 0.0, \
                  Mg_new,0.0, Si_new, S_new, Ar_new, Ca_new, Fe_new, 0.0, Ni56_new)
            HYD.write(hyd_line)
            ABN.write(abn_line)
    HYD.close()
    ABN.close()
    plt.figure(figsize=(11, 6))
    plt.plot(np.log10(hyd_[:,2]), np.log10(hyd_[:,3]),R_,np.log10(Rho_))
    plt.xlabel("log(R)")
    plt.ylabel(r"$log(\rho)$")
    plt.grid()
    plt.show()
    #plt.plot(R_,np.log10(T_))
    #plt.show()
```
